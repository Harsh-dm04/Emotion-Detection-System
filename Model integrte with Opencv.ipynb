```python
import cv2
import numpy as np

face_classifier=cv2.CascadeClassifier(r"C:\project\Facial Emotion Detection\haarcascade_frontalface_default.xml")

def face_detect(frame):

    faces=face_classifier.detectMultiScale(frame, 1.1,3,minSize=(30, 30))

    for (x, y, w, h) in faces:
    
        cv2.rectangle(frame, (x, y), (x + w, y + h), (200,0, 165), 2)


cap = cv2.VideoCapture(0)  

if not cap.isOpened():
    print("Error opening video stream or file")
    exit()

while True:
    ret,frame=cap.read()
    face_detect(frame)

    cv2.imshow("face_detection", frame)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model('Emotion_detection_model.keras')

# Define the emotion labels
emotion_labels = ['AngryğŸ˜ ', 'Disgust', 'FearğŸ˜¨', 'HappyğŸ™‚', 'SadğŸ˜”', 'SurpriseğŸ˜®', 'NeutralğŸ˜']



face_classifier=cv2.CascadeClassifier(r"C:\project\Facial Emotion Detection\haarcascade_frontalface_default.xml")

def face_detect(frame):

    faces=face_classifier.detectMultiScale(frame, 1.2,5,minSize=(30, 30))

    for (x, y, w, h) in faces:
    
         # Extract the region of interest (face)
        face = frame[y:y + h, x:x + w]
        
        # Convert the face to grayscale
        gray_face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

        # Resize the face to the required input size for the model (e.g., 48x48 for FER2013)
        resized_face = cv2.resize(gray_face, (48, 48))

        # Normalize the pixel values
        normalized_face = resized_face / 255.0

        # Reshape the face for the model (model expects input shape [1, 48, 48, 1] for grayscale)
        reshaped_face = np.reshape(normalized_face, (1, 48, 48, 1))

        # Make prediction using the model
        emotion_prediction = model.predict(reshaped_face)
        
        # Get the emotion with the highest probability
        emotion_label = np.argmax(emotion_prediction)

        # Define a list of emotion labels corresponding to the model's output
        emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

        # Display the predicted emotion on the frame
        cv2.putText(frame, emotion_labels[emotion_label], (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2)
        
        # Draw a rectangle around the face
        cv2.rectangle(frame, (x, y), (x + w, y + h), (200, 0, 165), 2)


cap = cv2.VideoCapture(0)  
if not cap.isOpened():
    print("Error opening video stream or file")
    exit()

while True:
    ret, frame = cap.read()
    face_detect(frame)

    cv2.imshow("face_detection", frame)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
